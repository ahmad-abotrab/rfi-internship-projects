# Plot 1: Line Plot

```python
# plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Port Name'], marker='o')
plt.title('Line Plot of Date vs Your Data')
plt.xlabel('Date')
plt.ylabel('Your Data')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Date'], df['Port Name'])
plt.title('Scatter Plot of Date vs Your Data')
plt.xlabel('Date')
plt.ylabel('Your Data')
plt.grid(True)
plt.show()
```

```python
port_count = df[0:10].groupby(['Date', 'Port Name']).size().unstack(fill_value=0)

# Plotting the bar plot
port_count.plot(kind='bar', stacked=True, figsize=(12, 6))

# Add labels and title
plt.title('Number of Ports per Year')
plt.xlabel('Year')
plt.ylabel('Number of Ports')
plt.legend(title='Port Name', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
